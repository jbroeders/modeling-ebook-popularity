```python
inputs = ['random', 'random', 'publish_year', 'publish_year', 'sentiment', 'sentiment', 'combined', 'combined']
metrics = ['MAE', 'RMSE', 'MAE', 'RMSE', 'MAE', 'RMSE', 'MAE', 'RMSE']
values = [0.339, 0.431, 0.300, 0.394, 0.277, 0.369, 0.262, 0.353]
```

```python
df = pd.DataFrame()
df['inp'] = inputs
df['Metric'] = metrics
df['val'] = values
```

```python
df
```

```python
ax = sns.catplot(
    data=df, kind="bar",
    x="inp", y="val", hue="Metric",
    ci="sd", alpha=.6, height=6
)

ax.despine(left=True)
ax.set_axis_labels("Input", "Error")
ax.legend.set_title("Metric")
ax.fig.suptitle('Random Forest performance')
plt.grid(axis='y')
```
